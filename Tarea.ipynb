```python
# Ejercicio 1: Mostrar el importe del IVA de un producto con un valor de 100€ y su precio final
precio_producto = 100
iva = 0.21  # IVA al 21%
importe_iva = precio_producto * iva
precio_final = precio_producto + importe_iva
print(f"El importe del IVA es: {importe_iva}€")
print(f"El precio final con IVA es: {precio_final}€")

# Ejercicio 2: Crear una variable numérica y comprobar sus rangos
numero = 15  # Puedes cambiar este valor para probar con diferentes números

if 0 <= numero <= 10:
    print("El número está entre 0 y 10.")
elif 11 <= numero <= 20:
    print("El número está entre 11 y 20.")
elif 21 <= numero <= 30:
    print("El número está entre 21 y 30.")
else:
    print("El número está fuera de los rangos especificados.")

# Ejercicio 3: Usar un while para mostrar los números del 1 al 100
print("Números del 1 al 100 con un while:")
contador = 1
while contador <= 100:
    print(contador)
    contador += 1

# Ejercicio 4: Mostrar los números del 1 al 100 con un for
print("Números del 1 al 100 con un for:")
for i in range(1, 101):
    print(i)
```
